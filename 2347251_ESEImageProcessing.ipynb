```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("image1.jpg")
plt.figure(figsize=(20,10))
plt.subplot(231), plt.imshow(img), plt.axis('off'), plt.title('Resized Image', size = 20)
```

```python
resize_img = cv.resize(img, (400, 400))

plt.figure(figsize=(20,10))
plt.subplot(231), plt.imshow(resize_img), plt.axis('off'), plt.title('Resized Image', size = 20)
```

```python
#grayscale conversion
gray_img = cv.cvtColor(resize_img, cv.COLOR_BGR2GRAY)

plt.figure(figsize=(20,10))
plt.subplot(231), plt.imshow(gray_img, cmap="gray"), plt.axis('off'), plt.title('Gray Scale Image', size = 20)
```

```python
# Rotate the cropped image by degrees
rows, cols = resize_img.shape
M = cv.getRotationMatrix2D((cols/2, rows/2), 45, 1)
rotated_roi = cv.warpAffine(resize_img, M, (cols, rows))

plt.figure(figsize=(20,10))
plt.subplot(231), plt.imshow(rotated_roi), plt.axis('off'), plt.title('Image Rotation', size = 20)
```
